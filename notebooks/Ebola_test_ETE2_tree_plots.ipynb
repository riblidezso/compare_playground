```python
import ete2
import numpy as np
```

##Firt glimpse at the tree

```python
#load it
mytree=ete2.Tree('ebola_phylo',format=1)

#make nodestyle: nothing
nstyle = ete2.NodeStyle()
nstyle["size"] = 0
for node in mytree.traverse():
    node.set_style(nstyle)
    
#set style
my_style = ete2.TreeStyle()
my_style.mode = "r" #rectangular
my_style.show_leaf_name=False #no 200 long names
my_style.rotation=90 #make it vertical
    
#draw it
mytree.render("%%inline",units="mm",w=600,tree_style=my_style)
```

#Color the nodes by country information
- information is in the name of the node
- leaves cluster by country

```python
mytree=ete2.Tree('ebola_phylo',format=1)

#collect labels
countries=set()
for leaf in mytree:
    countries.add( leaf.name.split('_')[4] )
    
#assign colors to countries
mycolors=['cyan','yellow','green','red','magenta','black','blue','tomato', 'palegreen', 'black', 
          'darkgrey', 'lawngreen', 'chocolate', 'crimson', 'forestgreen', 'slateblue', 'lightseagreen']
country_colors=dict([ [list(countries)[i],mycolors[i]] for i in xrange(len(countries)) ])

#set nodestyle to nothing
nstyle = ete2.NodeStyle()
nstyle["size"] = 0
for node in mytree.traverse():
    node.set_style(nstyle)
    
#color leaves
for leaf in mytree:
    leaf.add_face(ete2.CircleFace(2,country_colors[leaf.name.split('_')[4]] ,
                                  style="circle"),column=0,position='branch-right' )
    
#set style
my_style = ete2.TreeStyle()
my_style.mode = "r"
my_style.force_topology=True
my_style.show_leaf_name=False
my_style.rotation=90

#legend
for country in countries:
    my_style.legend.add_face(ete2.CircleFace(3, country_colors[country]), column=0)
    my_style.legend.add_face(ete2.TextFace('  '+country), column=1)
    my_style.legend_position=4
    
mytree.render("%%inline",units="mm",w=600,tree_style=my_style)
```

#Color the leaves by date

```python
mytree=ete2.Tree('ebola_phylo',format=1)

#collect labels
dates=set()
for leaf in mytree:
    dates.add('-'.join(leaf.name.split('_')[-3:-1]))

#assign continous colors to dates
date_colors=dict([ [sorted(dates)[i],'#%02X%02X%02X' % 
                    (255*i/float(len(dates)),0,255-255*i/float(len(dates)))] for i in xrange(len(dates)) ])


#set nodestyle to nothing
nstyle = ete2.NodeStyle()
nstyle["size"] = 0
for node in mytree.traverse():
    node.set_style(nstyle)

#color leaves
for leaf in mytree:
    leaf.add_face(ete2.CircleFace(2,date_colors['-'.join(leaf.name.split('_')[-3:-1])] ,
                                  style="circle"),column=0,position='branch-right' )
    
#set style
my_style = ete2.TreeStyle()
my_style.mode = "c"
my_style.force_topology=True
my_style.root_opening_factor=0.8
my_style.show_leaf_name=False
my_style.rotation=90 

for date in sorted(date_colors.keys()):
    my_style.legend.add_face(ete2.CircleFace(3, date_colors[date]), column=0)
    my_style.legend.add_face(ete2.TextFace('  '+date), column=1)

mytree.render("%%inline", h=600, units="mm", tree_style=my_style)
```

##Plot the districts

```python
mytree=ete2.Tree('ebola_phylo',format=1)
for node in mytree.traverse():
    node.dist=np.min([1000*node.dist,1],axis=0)

#collect labels
districts=set()
for leaf in mytree:
    districts.add(leaf.name.split('_')[5])

#assign colors to districts
mycolors=['aquamarine','yellow','green','purple','magenta','black','blue','tomato', 'red', 'black', 
          'darkgrey', 'lawngreen', 'chocolate', 'salmon', 'forestgreen', 'palegreen', 'skyblue',
          'gold', 'olive','cyan']
dist_colors=dict([ [list(districts)[i],mycolors[i]] for i in xrange(len(districts)) ])
    
#set style to nothing
nstyle = ete2.NodeStyle()
nstyle["size"] = 0
for node in mytree.traverse():
    node.set_style(nstyle)
    
    
#color leaves
for leaf in mytree:
    leaf.add_face(ete2.CircleFace(2,dist_colors[leaf.name.split('_')[5]] ,
                                  style="circle"),column=0,position='branch-right' )

my_style = ete2.TreeStyle()
my_style.mode = "c"
my_style.root_opening_factor=0.8
my_style.show_leaf_name=False
my_style.arc_start=0

for district in sorted(dist_colors.keys()):
    my_style.legend.add_face(ete2.CircleFace(3, dist_colors[district]), column=0)
    my_style.legend.add_face(ete2.TextFace('  '+district), column=1)

mytree.render("%%inline", h=600, units="mm", tree_style=my_style)
```
